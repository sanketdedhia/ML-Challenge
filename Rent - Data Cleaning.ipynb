```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import missingno as msno
```

```python
data = pd.read_csv('housing_train.csv')
data.head()
```

# Data Cleaning

```python
data.describe()
```

Dropping 4 columns as they seem irrelevant.

```python
data.drop(['region_url','description','image_url','url'], axis = 1, inplace = True)
```

```python
data.head()
```

```python
data.shape
```

CHecking for NaN/null values

```python
pd.isna(data).sum()
```

lat, long and state seems to be missing values for very few rows. These can be dropped.

```python
data.dropna(subset = ['lat','long','state'], inplace = True)
```

```python
pd.isna(data).sum()
```

Replacing missing parking_optoins and laundry_options values by their mode

```python
data.parking_options.fillna(data.parking_options.mode()[0], inplace = True)
```

```python
pd.isna(data).sum()
```

```python
data.laundry_options.fillna(data.laundry_options.mode()[0], inplace = True)
```

```python
pd.isna(data).sum()
```

```python
 data.select_dtypes('object')
```

Dropping region column for 2 reasons.
1 - Since we have the latitudes, longitudes and states columns, region seems to be irrelevant.
2 - There is no consistency in the region column. The values are a mix of states, county, city.

```python
data.drop(['region'], axis = 1, inplace = True)
```

```python
data.drop(['id'], axis = 1, inplace = True)
```

```python
data
```

```python
cat_var = data.select_dtypes('object').columns
```

```python
cat_var
```

Getting dummies for categorical variables

```python
x = pd.get_dummies(data[cat_var], drop_first=True)
```

```python
x.columns
```

```python
data_final = pd.concat([data,x], axis = 'columns')
```

```python
data_final
```

```python
data_final.drop(cat_var, axis = 1, inplace = True)
```

```python
data_final.head()
```

```python
data_final.describe()
```

There are a lot of unrealistic values in sqfeet, beds and baths columns. Cleaning those below.

```python
data_final[data.sqfeet > 5000]
```

```python
data_final = data_final[data_final.beds < 6]
data_final = data_final[data_final.baths < 5]
data_final = data_final[data_final.price < 10000]
data_final = data_final[data_final.sqfeet < 5000]
```

```python
data_final = data_final[data_final.beds > 0]
data_final = data_final[data_final.baths > 0]
data_final = data_final[data_final.price > 0]
data_final = data_final[data_final.sqfeet > 400]
```
